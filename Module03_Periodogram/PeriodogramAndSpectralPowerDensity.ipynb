## 1. Load the required libraries, load the data, and double-check it

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig

# Load the Henry's fork discharge dataset
df = pd.read_pickle('HenrysForkDischarge_WY2001-2015.pkl')

# Print the first and last 10 records just to make sure we loaded the data okay
qrows = np.concatenate([np.arange(0,10,1),np.arange(-11,-1,1)])
df.iloc[qrows,:]
```

```python
Q = df['Q'].values

f, Pxx_den = sig.periodogram(Q,1.0/24.0)

plt.figure(figsize=(14,10))
plt.xlabel('Frequency [hr${}^{-1}$]', fontsize=16)
plt.ylabel('Spectral Power', fontsize=16)
plt.semilogy(f[1:-1], Pxx_den[1:-1])
```

```python

f, Pxx_den = sig.welch(Q,1.0/24.0)

plt.figure(figsize=(14,10))
plt.xlabel('Frequency [hr${}^{-1}$]', fontsize=16)
plt.ylabel('Spectral Power', fontsize=16)
plt.semilogy(f[1:-1], Pxx_den[1:-1])
```

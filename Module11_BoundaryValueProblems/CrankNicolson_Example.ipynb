```python
import numpy as np
import matplotlib.pyplot as plt

K = 2.0

xL = 0.0
xR = 100.0
dx = 1

Nx = (xR - xL)/dx + 1

ti = 0.0
tf = 500.0
dt = 0.1

Nt = (tf - ti)/dt + 1

hi = 8.0
hf = 4.0

hL = np.linspace(hi,hi,num=Nt,endpoint=True)
hR = np.linspace(hi,hf,num=Nt,endpoint=True)

x = np.linspace(xL,xR,num=Nx,endpoint=True)

h0 = hi*np.ones(shape=(int(Nx),1))
```

```python
alpha = K*dt/dx**2
print("alpha = ",alpha)
print("dt = ",dt)
print("dx = ",dx)
```

```python
# Construct A matrix and its inverse
A = np.matrix(np.diagflat(1+2.0*alpha*np.ones((int(Nx),1)),k=0) + np.diagflat(-alpha*np.ones((int(Nx-1),1)),k=1) \
    + np.diagflat(-alpha*np.ones((int(Nx-1),1)),k=-1))

A[0,:] = np.zeros(shape=(1,int(Nx)))
A[0,0] = 1.0
A[-1,:] = np.zeros(shape=(1,int(Nx)))
A[-1,-1] = 1.0

Ainv = np.linalg.pinv(A)

B = np.matrix(np.diagflat((1-2.0*alpha)*np.ones((int(Nx),1)),k=0) + np.diagflat(alpha*np.ones((int(Nx-1),1)),k=1) \
    + np.diagflat(alpha*np.ones((int(Nx-1),1)),k=-1))

B[0,:] = np.zeros(shape=(1,int(Nx)))
B[0,0] = 1.0
B[-1,:] = np.zeros(shape=(1,int(Nx)))
B[-1,-1] = 1.0
```

```python
H = np.matrix(np.zeros(shape=(int(Nx),int(Nt))))

plt.figure(figsize=(6,4), dpi=150)
plt.xlabel("Distance")
plt.ylabel("Pressure head")
plt.title("Example 1D groundwater model: Implicit method")
plt.ylim(0.0,hi+2.0)
plt.plot(x,h0,'b',linewidth=0.5)

H[:,0] = np.matrix(h0[:,0]).T

for i in range(1,int(Nt)):
    hinit = H[:,i-1]
        
    hnext = Ainv*B*hinit 
    hnext[0] = hL[i]
    hnext[-1] = hR[i]

    H[:,i] = hnext[:,0]
    
    if (i % 200) == 0:
        plt.plot(x,H[:,i],'b',linewidth=0.5)
        
plt.show()
```

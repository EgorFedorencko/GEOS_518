```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('HenrysForkDischarge.csv', names=['Y','M','D','Q'], skiprows=1)
```

```python
pdDates = pd.to_datetime(df.Y*10000+df.M*100+df.D,format='%Y%m%d')

df['SerDates'] = pdDates
```

```python
df.plot(x='SerDates', y='Q',figsize=(14,10))
```

```python
Q = df['Q'].values
Qt   = Q[1:-1]
Qtm1 = Q[0:-2]

plt.figure(figsize=(14,10))
plt.plot(Qtm1,Qt,'o')
plt.xlabel('$Q_{t-1}$ m$^3$/s')
plt.ylabel('$Q_t$ m$^3$/s')
plt.show()
```

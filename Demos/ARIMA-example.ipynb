# Autoregressive Integrated Moving Average (ARIMA) example

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt

df = pd.read_csv('weekly_in_situ_co2_mlo.csv',header=0)
```

```python
DateArray = df.as_matrix()
DateArray = DateArray[:,0:3].astype(int)

SerDates = [dt.datetime(*x) for x in DateArray]

df['SerDates'] = SerDates
```

```python
CO2 = df['CO2'].values
dCO2 = CO2[1:-1] - CO2[0:-2] # First derivative
d2CO2 = dCO2[1:-1] - dCO2[0:-2] # Second derivative
d3CO2 = d2CO2[1:-1] - d2CO2[0:-2] # Third derivative
```

```python
plt.figure(figsize=(14,10))
plt.subplot(411)
plt.plot(CO2)
plt.subplot(412)
plt.plot(dCO2)
plt.subplot(413)
plt.plot(d2CO2)
plt.subplot(414)
plt.plot(d3CO2)
plt.show()
```
